```python
#Average of Price
average_price = mycol.aggregate([{"$group":{"_id": 0, "avg_price":{"$avg":"$price"}}}])
for result in average_price:
  print("average price of Airbnb listings: $", result['avg_price'])

# standard deviation for price - to quantifies the variance or dispersion in dataset
Std_deviationPrice = mycol.aggregate([{"$group":{"_id":0, "std_dev_price":{"$stdDevPop":"$price"}}}])
for res in Std_deviationPrice:
  print("Standard deviation for Price : $", res['std_dev_price'])
```

```python
#Descriptive statistics for review score
#desc_stat = mycol.aggregate([{"$group":{"_id":0, "count":{"$sum":1}, "mean_reviewScore":{"$avg":"$review_scores"}, "median_reviewScore":{"$median":"$review_scores"}, "std_dev_reviewScore":{"$stdDevPop":"$review_scores"}}}])
pipeline = [{"$group":{"_id":0, "count":{"$sum":1}, "mean_reviewScore":{"$avg":"$number_of_reviews"}, "std_dev_reviewScore":{"$stdDevPop":"$number_of_reviews"}}}]
desc_stat = mycol.aggregate(pipeline)
result = list(desc_stat)

if result:
    stats = result[0]
    print("Descriptive Statistics for Number of Reviews")
    print("Count: ",stats["count"])
    print("Mean of Review Score: ",stats["mean_reviewScore"])
   # print("Median of Review Score: ",stats["median_reviewScore"])
    print("Standard Deviation of Review Score: ",stats["std_dev_reviewScore"])
else:
    print("Data Does Not Exist")
                                        
```
